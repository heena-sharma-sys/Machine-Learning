Multi-Class Classification on MNIST dataset.
The MNIST database of handwritten digits has a training set of 60,000 examples, and a test set of 10,000 examples with each image of size 28 * 28. Each pixel of image contains value 0-255. 

```python
import pandas as pd
import numpy as np
import tensorflow
from tensorflow import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[110])
plt.show()
```

```python
# Now scale all image values from 0-1 
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
#Lets flattened the data from 2-d to 1-d so that we can easily 
#apply data to layers of 28*28=784 neurons
```

Now let's train the DL model, here we create **Sequential model** and add layers with 28 * 28 = 784 nodes. 

```python
from keras.models import Sequential 
from tensorflow.keras.layers import Dense, Flatten
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu')) #hidden layer
model.add(Dense(10, activation='softmax')) #output layer with 10 output classes
```

```python
model.summary()
#In first hidden layer: 
#(784 inputs * 128 neurons) + 128 bias/neuron = 100480 parameters

#In output layer:
# 128 inputs (or ourpurs from input layer) *10 neuron + 10 bias/neuron = 1290

# after adding up everything, the total number of tunable parapeters: 
#100480 + 1290= 101770
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
hist = model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
y_prediction = y_pred.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_prediction, y_test)
```
